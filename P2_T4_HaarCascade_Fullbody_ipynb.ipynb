```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

fullbody_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/IA/HaarCascade/haarcascade_fullbody.xml');
img = cv2.imread("/content/drive/MyDrive/IA/Images/cuepoc.jpg")
fullbody = fullbody_cascade.detectMultiScale(img,scaleFactor=2, minNeighbors=1)
print(fullbody)
idx=0
for (sx,sy,sw,sh) in fullbody:
  cv2.rectangle(img,(sx,sy),((sx+sw),(sy+sh)),(0,255,0),2)
  
cv2_imshow(img)
cv2.imwrite('/content/drive/MyDrive/IA/Images/cuepoc_detacadas.jpg',img)
```
