##Ejercicio 3:Vector con lista maximo y minimo

```python
matriz = [
    [103, 23, 5, 7, 10],  
]
for lista in matriz:
    maximo = max(lista)
    minimo = min(lista)
    print(f"El maxmimo es {maximo} \nEl minimo es {minimo}")
```
